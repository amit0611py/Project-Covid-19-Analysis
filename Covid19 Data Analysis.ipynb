```python
# useless columns
df=cd.drop(["Lat","Long"],axis=1)
df
```

# Aggregating the rows by the country

```python
df=df.groupby("Country/Region").sum()
df
```

# find maxmimum infection rate

```python
df.loc["India"].diff().max()
```

# find maximum infection rate for all of the countries

```python
countries=list(df.index)
max_infected_rates=[]
for c in countries:
    max_infected_rates.append(df.loc[c].diff().max())
```

```python
max_infected_rates
```

```python
df["max_infected_rates"]=max_infected_rates
df
```

# create a new dataframe with only needed column

```python
corona_data=pd.DataFrame(df["max_infected_rates"])
corona_data
```
